```python
# data set
# input
x_data = [1, 2, 3, 4, 5]
# output
y_data = [2, 4, 6, 8, 10]
```

# Start with a random weight

# start with a random weight

```python
w = 1.0
```

# Forward function 

```python
def forward(x):
    return x * w
```

```python
def loss(x, y):
    # take prediction for input
    y_pred = forward(x)
    return (y_pred-y) * (y_pred-y)
```

```python
import numpy as np

# take weight list and mse list to show in graph
w_list = []
mse_list = []

for w in np.arange(0.0, 4.1, 0.1):
    print("W:", w)
    l_sum = 0
    for x, y in zip(x_data, y_data):
        y_pred = forward(x)
        l = loss(x, y)
        l_sum += l
        print("\t", x, y, y_pred, l)
    
    print("Mse:", l_sum/5)
    w_list.append(w)
    mse_list.append(l_sum/5)
```

```python
import matplotlib.pyplot as plt

plt.plot(w_list, mse_list)
plt.xlabel("W")
plt.ylabel("Loss")
plt.show()
```

```python
# This is the way to find the best weight
```

```python
# take weight list and mse list to show in graph
w_list = []
mse_list = []
min_mse = np.inf
best_w = 0
for w in np.arange(0.0, 4.1, 0.1):
    print("W:", w)
    l_sum = 0
    for x, y in zip(x_data, y_data):
        y_pred = forward(x)
        l = loss(x, y)
        l_sum += l
        # print("\t", x, y, y_pred, l)

    mse = l_sum / 5
    print("Mse:", mse)
    w_list.append(w)
    mse_list.append(l_sum / 5)
    if mse < min_mse:
        min_mse = mse
        best_w = w

print(best_w)
w = best_w
```
